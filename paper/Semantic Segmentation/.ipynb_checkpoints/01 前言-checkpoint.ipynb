4. 常见语义分割评价指标
$$
\begin{array}
\hline \begin{array}
\text { Pixel Accuracy } \\
\text { (Global Acc) }
\end{array} & \frac{\sum_i n_{i i}}{\sum_i t_i} \\
\hline \text { mean Accuracy } & \frac{1}{n_{c l s}} \cdot \sum_i \frac{n_{i i}}{t_i} \\
\hline \text { mean IoU } & \frac{1}{n_{c l s}} \cdot \sum_i \frac{n_{i i}}{t_i+\sum_j n_{j i}-n_{i i}}\\
\end{array}
$$
- $n_{i j}$: 类别$i$被预测成类别$j$的像素个数
- $n_{c l s}$: 目标类别个数（包含背景）
- $t_i = \sum_i n_{i j}$: 目标类别$i$的总像素个数（真实标签）
